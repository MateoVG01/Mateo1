```python
df = pd.read_csv('datasets/airline_data.csv',
                        low_memory=False)
df.tail(2)
```

```python
test1=df.loc[:,['DepDelay','DepartureDelayGroups']]
test2=df['DepartureDelayGroups'].min()
test2
```

```python
test1=df.loc[:,['DepDelay','DepartureDelayGroups']]
test2=df['DepartureDelayGroups'].max()
test2
```

```python
test3=df.groupby(['DepDelay','DepartureDelayGroups']).size()
test3
```

```python
pd.pivot_table(df,
              index='Year'
             )
```

```python
my_pivot = pd.pivot_table(df,
              index=['Year','Quarter']
             )
my_pivot.head(12)
```

```python
my_pivot = pd.pivot_table(df,
              index=['Year','Quarter'],
                          columns='DepartureDelayGroups',
                          aggfunc='sum',
                          values='DepDelay'
             )
my_pivot.head(12)
```

```python
my_pivot = pd.pivot_table(df,
              index=['Year','Quarter'],
                          columns='DepartureDelayGroups',
                          aggfunc='sum',
                          values='DepDelay',
                          margins=True
             )
my_pivot.head(12)
```
