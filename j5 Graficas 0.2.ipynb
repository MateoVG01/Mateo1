```python
x = np.arange(0, 3*np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y)
```

```python
ysin = np.sin(x)
ycos = np.cos(x)

plt.plot(x, ysin)
plt.plot(x, ycos)

plt.xlabel('x axis label')
plt.ylabel('y axis label')

plt.title('sine & cosine Funcions')
```

```python
#2 rows, 1 colums, position 1

plt.subplot(2,1,1)  
plt.plot(x,ysin)
plt.title('Sine')

plt.subplot(2,1,2)  
plt.plot(x, ycos)
plt.title('Cosine')
```

```python
path = 'datasets/Cartwheeldata.csv'
d1f = pd.read_csv(path)
d1f.head(2)
```

```python
sns.lmplot(x = 'Wingspan',
           y = 'Height',
           data = d1f
          )
```

```python
sns.lmplot(x = 'Wingspan',
           y = 'Height',
           data = d1f,
           fit_reg = False,
           hue = 'Gender'
          )
plt.title('Scatter plot of Wingspan vs Height')
```

```python
sns.swarmplot(x = 'Gender',
              y = 'Wingspan',
              data = d1f,
             )
plt.title('Gender vs Wingspan, the Swarmplot')
```

```python
sns.boxplot(data = d1f.loc[:,'Age'])
```
